Consequences for false positives:
- Some people who were not in churn will be included into the loyalty measure -> means more expanses than necessary
Consequences for false negatives:

## Using standard False positive/False negative ratio

```python
churn_1_TP = 1978 
churn_1_FP = 972
churn_1_FN = 4643 
churn_1_TN = 10317 

churn_2_TP = 2050
churn_2_FP = 1272
churn_2_FN = 4343 
churn_2_TN = 10245 


accuracy_1 = (churn_1_TP + churn_1_TN) / (churn_1_TP + churn_1_TN + churn_1_FN + churn_1_FP)
accuracy_2 = (churn_2_TP + churn_2_TN) / (churn_2_TP + churn_2_TN + churn_2_FN + churn_2_FP)
accuracy_1, accuracy_2
```

```python
precision_1 = churn_1_TP / (churn_1_TP + churn_1_TN)
precision_2 = churn_2_TP / (churn_2_TP + churn_2_TN)

precision_1, precision_2
```

```python
recall_1 = churn_1_TP / (churn_1_TP + churn_1_FN)
recall_2 = (churn_2_TP) / (churn_2_TP + churn_2_TN)

recall_1, recall_2
```

```python
specificity_1 = 1 - recall_1
specificity_2 = 1 - recall_2

specificity_1, specificity_2
```

## Using F1 score

```python
f1_1 = (2 * precision_1 * recall_1) / (precision_1 + recall_1)
f1_2 = (2 * precision_2 * recall_2) / (precision_2 + recall_2)

f1_1, f1_2
```
